Atılan 2 zarın toplamının ihtimal analizi

```python
import random
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

def roll_toplam():
    zar1=random.randrange(1,7)
    zar2=random.randrange(1,7)
    return zar1+zar2

cikti=[roll_toplam() for i in range(360000)]
values , frequencies=np.unique(cikti,return_counts=True) 

                                    
sns.set_style("whitegrid")
axes=sns.barplot(x=values,y=frequencies,palette="bright")

axes.set_title(f"{len(cikti)} kez atılan 2 zar toplamı!")

axes.set(xlabel="2 Zar toplamı", ylabel="Atış sayısı")

axes.set_ylim(top=max(frequencies)*1.20)

for bar,frequency in zip(axes.patches,frequencies):
    text_x=bar.get_x()+bar.get_width()/2.0
    text_y=bar.get_height()
    text=f"{frequency:,}\n{frequency/len(cikti):.1%}"
    
    axes.text(text_x,text_y,text,
             fontsize=13,ha="center",va='bottom')

plt.show()    
```

Craps oyununun kazanma ve kaybetme oranlarının gösterilmesi

```python

def craps():
    
    def zar_atmak():
   
        zar1=random.randrange(1,7)
        zar2=random.randrange(1,7)  
        return (zar1,zar2)  ##zarları paketleme

    def zar_goster(zar):
        zar1,zar2=zar
        #print(f'Kullanıcı atışı {zar1} + {zar2} = {sum(zar)}') #değerleri paketten çıkarır

    zar_degeri = zar_atmak()
    zar_goster(zar_degeri)
    zar_toplam=sum(zar_degeri)
    counter=0
    if zar_toplam in(7,11):
        sonuc = "kazandınız"
        counter=1
    elif zar_toplam in(2,3,12):
        sonuc = "kaybettiniz"
        counter=1
    else :
        sonuc = "Devam"
        puan=zar_toplam
        #print("Puanınız : ",puan)
        counter+=1
    
    while sonuc == "Devam":
        zar_degeri= zar_atmak()
        zar_goster(zar_degeri)
        zar_toplam=sum(zar_degeri)
        counter+=1
        if zar_toplam==puan:
            sonuc="kazandınız"
        elif zar_toplam==7:
            sonuc="kaybettiniz"
    
    return sonuc


sonuclar=[craps() for i in range(100000)]

values,frequencies = np.unique(sonuclar,return_counts=True)

sns.set_style("whitegrid")
axes=sns.barplot(x=values,y=frequencies,palette="bright")

axes.set_title(f"{len(sonuclar)} kez oyananınca ihtimaller")

axes.set_ylim(top=max(frequencies)*1.10)

for bar,frequency in zip(axes.patches,frequencies):
    
    text_x = bar.get_x() + bar.get_width() / 2.0 
    text_y = bar.get_height() 
    text = f'{frequency:,}\n{frequency / len(sonuclar):.1%}'
    axes.text(text_x, text_y, text, 
        fontsize=11, ha='center', va='bottom')




```
