```python
import random
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

def roll_toplam():
    zar1=random.randrange(1,7)
    zar2=random.randrange(1,7)
    return zar1+zar2

cikti=[roll_toplam() for i in range(360000)]
values , frequencies=np.unique(cikti,return_counts=True) 

                                    
sns.set_style("whitegrid")
axes=sns.barplot(x=values,y=frequencies,palette="bright")

axes.set_title(f"{len(cikti)} kez atılan 2 zar toplamı!")

axes.set(xlabel="2 Zar toplamı", ylabel="Atış sayısı")

axes.set_ylim(top=max(frequencies)*1.20)

for bar,frequency in zip(axes.patches,frequencies):
    text_x=bar.get_x()+bar.get_width()/2.0
    text_y=bar.get_height()
    text=f"{frequency:,}\n{frequency/len(cikti):.1%}"
    
    axes.text(text_x,text_y,text,
             fontsize=13,ha="center",va='bottom')

plt.show()    
```
